```python
# 

import pandas as pd
import numpy as np

filename = 'E:/dataAnalysisFile/24064925aueh/DataAndCode/chapter5/demo/data/sales_data.xls'
data = pd.read_excel(filename, index_col=u'') #

# 
# 1 0
data[data==u''] = 1
data[data==u''] = 1
data[data==u''] = 1
data[data != 1] = 0
x = data.iloc[:, :3].values.astype(int)
y = data.iloc[:, 3].values.astype(int)

from keras.models import Sequential
from keras.layers.core import Dense, Activation


model = Sequential() #
model.add(Dense(input_dim = 3, units = 10))
model.add(Activation('relu')) #relu
model.add(Dense(input_dim = 10, units = 1))
model.add(Activation('sigmoid')) #0-1sigmoid
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
#binary_crossentropybinary
#mean_squared_errorcategorical_crossentropy
#adamsgdrmsprop
model.fit(x, y, epochs = 1000, batch_size = 10) #
yp = np.argmax(model.predict(x),axis=1).reshape(len(y)) #
from cm_plot import * #
cm_plot(y,yp).show() #
```
